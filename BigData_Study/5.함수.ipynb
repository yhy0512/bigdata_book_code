### 나. 디폴트(default) 매개변수
- 매개변수 기본값(default) 설정으로 인수와 매개변수의 개수가 맞지 않아도 설정된 기본값으로 함수의 동작을 실행
- 함수를 정의할 때 매개변수의 기본값을 지정 




```python
def print_line3(symbol='x',num =3): # 매개변수의 기본값 지정하면 인수를 생략해도 실행 가능
    for i in range(1,num + 1) :
        print(symbol*i)
```

### 라.가변(variable length) 인수
- 인수의 개수에 상관없이 함수를 호출
- 함수의 매개변수 앞에 ‘*’로 정의


```python
def add(*numbers) :  #매개변수의 개수가 정해지지 않음. 인수의 개수에 상관없이 호출
    result = 0
    for x in numbers:
        result += x
    print(result)  
    
```

```python
add()
```

```python
add(10, 20)
```

```python
add (10,20,30,40,50)
```

### 마. 결과값 반환

* 지역변수 : 함수 내에서 사용되는변수
* 전역변수 : 함수 바깥에서 정의한 변수 
* 지역변수를 전역변수로 바꾸는 명령어 : grobal


```python
def add(a,b):
    total = a+b  # 지역변수 total 생성

total = 0        # 전역변수 total 생성   
add(10,20)
print(total)     # 전역변수 total 출력  
```

```python
def add(a,b):
    global total # 전역변수 total 사용 선언
    total = a+b  

total = 0        # 전역변수 total 생성   
add(10,20)
print(total)     # 전역변수 total 출력  
```


* 함수의 반환값 : 결과로 나오는 반환값,  return 사용

## 3.클래스(class)
- 함수를 반복해서 사용할때 전역변수로 인해 반복이 안될경우 사용이 쉽게 하기 위한 방법
 

```python
# 더하기를 하는 함수(결과를 전역변수로 반환)

def add(num):
  global result
  result = result + num
  return result



result = 0

print(add(3))
```

```python
print(add(4))
```

```python
# result1 = 3+4, result2 = 3+7 을 뽑아 내려면

result = 0
print(add(3))
print(add(4))
### 코드
result = 0   #초기화를 안한다면 앞의 결과가 누적됨.
print(add(3))
print(add(7))

```

image.png

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```
