#2.출력하기 
 - print(파라미터)
 - 파라미터는 변수 또는 문자 ,숫자
 - 괄호안에 문자(출력할 글자)를 입력할 때에는 따옴표안에  넣어야함.예) print(“Hello python”)
 - 문자형에서 이스케이프를 사용하면 바로 다음에 오는 기호를 무시하면 아래와 같이 나타남.
  - \t : 키보드에서 Tab 을 클릭한 거리 만큼 글자사이의 간격을 띄어줌.
  - \n : 줄바꿈
  - \ \ : \를 표현 


```python
print('Hello World')
```

```python
print('Hello \'World\'')
```

```python
print('Hello \tWorld')
```

```python
print('Hello \nWorld')
```

```python
print('Hello \a World')
```

```python
print('Hello \bWorld')
```

```python
name = "홍길동" 
age = 34
height = 178.538
print("이름은 %s, 나이는 %d, 키는 %f 입니다."
      %(name,age,height))
```

```python
print("이름은 %s, 나이는 %d, 키는 %6.2f 입니다."
      %(name,age,height))
```

```python
print("이름은 {0}, 나이는 {1}, 키는 {2} 입니다.".
      format(name,age,height))
```

```python
print("이름은 {0}, 키는 {2}, 나이는 {1} 입니다."
      .format(name,age,height))
```

```python
print("이름은 {name}, 키는 {height}, 나이는 {age} 입니다."
      .format(name="김철수",age=8,height=132.75))
```

```python
print("이름은 {name}, \n키는 {height}, \n나이는 {age} 입니다."
      .format(name="김철수",age=8,height=132.75))
```
