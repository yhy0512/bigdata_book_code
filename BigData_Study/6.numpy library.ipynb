- random 메소드에 입력값으로 5를 입력해주면 0에서 1 사이의 무작위의 수 5개를 생성

- all(): 배열의 모든 원소가 조건을 만족하면 True를 출력, 그렇지 않으면 False를 출력

- any(): 배열에 조건을 만족하는 원소가 있으면 True를 출력


```python
ary = np.random.random(5)
print(ary)
print(np.all(ary >= 0.3))
print(np.any(ary >= 0.3))
```

- linspace() : 시작과 끝의 값을 지정하고 그 사이의 값을 균등하게 배분

- retstep=True이면 샘플 간의 간격을 출력

- endpoint=False이면 마지막 값을 제외하고 샘플값을 만들어냄


```python
# 등간격 나누기

np.linspace(0,12,4)
```

```python
np.linspace(0,12,4,retstep=True)
```

```python
np.linspace(0,12,4, endpoint=False,retstep =True)
```

- logspace() :
10에서 100까지 5개의 대수(로그) 형태의 간격으로 나누어진 원소들을 배열로 생성

```python
np.logspace(np.log10(10),np.log10(100),5)
```

```python
np.logspace(1,2,5)
```

- 먼저 리스트를 Numpy 형식으로 변환 후 슬라이싱 해야 함
- 리스트의 인덱스는 반드시 정수나 슬라이스여야 함

- 파이썬의 2차원 리스트에서 모든 원소의 두 번째 원소들만 출력
image.png

```python
list1 = [[1, 11], [2, 12], [3, 13]]
print(list1)
print(list1[:][1])  # 의도하지 않은 결과 출력
print(list1[:, 1])  # error발생
```
