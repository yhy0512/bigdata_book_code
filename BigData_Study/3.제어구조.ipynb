# 2.조건문 
- 가장 단순한 형태의 조건문으로, 조건식이 True인 경우 실행할 명령을 기술
- if 뒤에 조건식을 기술한 다음, 반드시 콜론(:)을 입력
- 조건식이 참인 경우 실행할 동작은 들여쓰기 후 입력

#### **if 문**의 콜론(:) 뒤에서 Enter 를 입력하면 자동으로 **들여쓰기**가 되며, Tab 이나 Space bar 로 직접 들여쓰기를 설정.


## 가. if문

```python
x= int(input("입력:"))
if x > 0 :
  print("양수입니다.")
print("x의 값은 ",x)
```

##나. if ~ else 문
- 조건식 검사 결과가 참일 때와 거짓일 때, 각각 실행할 명령이 있는 경우에 사용


```python
x= int(input("입력:"))
if x > 0 :
  print("양수입니다.")
else:
  print("양수가 아님.")
```

## 다. 다중if 문


```python
x= int(input("입력:"))
if x > 0 :
  print("양수입니다.")
elif x == 0:
  print("0 입니다.")
else:
  print("음수입니다.")
print("x의 값은 ",x)
```

### 문제3) 올림픽에서 양궁 점수가 한국 120점, 미국 117점, 중국 118점을 획득했습니다. 점수가 높은순으로 금, 은,동이 정해집니다. 한국은 어느 메달일까요? 프로그램이 마지막에 끝이라고 출력한다.

image.png

```python
KOR = 120
USA = 117
CHN = 118

if (KOR < USA and KOR < CHN):
    print("동메달")
elif (KOR > USA and KOR > CHN):
    print("금메달")
else:  
    print("은메달")
print ("끝")
```

### 문제5) 온도 (화씨 온도를 섭씨 온도로 변환), 무게(파운드를 킬로그램으로 변환), 길이(인치를 센티미터로 변환) 단위변환기를 만드시오.
- 어떤 단위인지와 변환할 값을 입력 받고 변환결과는 단위와 함께 출력하시오.

image.png

```python
menu = int(input("번호 입력(1:온도, 2:무게, 3:길이) : "))
value = float(input("변환할 값 입력 : "))

if menu == 1 :      # 온도 단위 변환
    print("변환 결과 = %.2f ℃" % ((value - 32) / 1.8)) 
elif menu == 2 :    # 무게 단위 변환
    print("변환 결과 = %.2f ㎏" % (value * 0.453592)) 
elif menu == 3 :    # 길이 단위 변환
    print("변환 결과 = %.2f ㎝" % (value * 2.54)) 
else :              # 잘못된 메뉴 선택
    print("메뉴 선택 오류!")
```

# 3.반복문
 - 반복문 : 지정한 범위에 해당하는 만큼 명령문을 반복해서 실행

 - 반복문의 흐름을 제어하는 break와 continue를 사용하여 출력 패턴에 알맞은 프로그램을 작성




##가.for 문

- for문: 그룹 자료형을 반복 범위로 지정하여 사용하는 것으로 그룹 자료형의 원소를 순서대로 반복 변수에 대입하고 명령문을 실행, 반복범위에 있는 원소의 개수만큼 반복해서 수행
- for 문은 특정 횟수만큼 반복하기 위한 문장으로, range( ) 함수와 함께 많이 사용

image.png



### 1) range사용
- range( ) 함수의 인수로 사용하는 정수가 반복 횟수가 되고, 반복할 문장들을 들여쓰기로 입력


image.png



- range( ) 함수는 정해진 구간의 정수들(수열)을 생성하는 함수
- range(5)인 경우 0~4까지 5개의 값을 만들 수 있고, 반복이 실행될 때마다 값을 하나씩 반환



```python
for x in range(0,5,1) : 
  print(x)
```

* 예제 1) 0부터 10까지 정수를 차례대로 더한 값을 출력하는 프로그램을 for 문을 사용하여 만들어보시오.
 * 출력 예

 0 1 3 6 10 15 21 28 36 45 55


```python
s = 0   # 초기화

for x in range(11): 
    s = s + x
   # s += x   # 위와 동일  
    print(s,end = " ")
```

### 2) 그룹형 자료형 사용

* 예제 2)  test_list 에 있는 것을 순서대로 뒤에 ! 를 붙여서 출력하시오.

```python
test_list = ['one', 'two', 'three']
for i in test_list:
    x = i + '!'
    print(x) 
```

* 예제 3) 5명이 시험을 본 결과가 [90, 25, 67, 45, 93]으로 나왔습니다.  시험점수가 60이상이면 합격입니다. 

  **합격 불합격을 1번 학생부터 5번 학생까지 출력해주세요.** 

```python
number = 0  #초기화
for score in [90, 25, 67, 45, 93]:
    number += 1 
    if score >= 60:
        print("%d번 학생은 합격입니다." %number)
    else: 
        print("%d번 학생은 불합격입니다." %number) 
```

## 나. while 문
- While문 :조건식의 결과가 참인 동안 명령문을 반복해서 수행
image.png

* 예제 1) 1번째 줄은  * 한개, 2번째 줄은 * 2개, ~ 5번째줄은 * 5개를 출력하시오.

image.png

```python
i = 0  # 초기화
while i < 5:
    i += 1
    print('*' * i) 
```

### 문제 6)0부터 10까지 정수를 차례대로 더한 값을 출력하는 프로그램을 while 문을 사용하여 만들어보시오.




image.png

```python
x = 0
s = 0
while x <= 10:
    s += x
    print(s, end = " ")
    x += 1 
```

```python
s = 0       # 합계를 저장하기 위한 변수

for x in range(0, 11):  # 0부터 10까지 반복
    s += x  # s = s + x와 동일, s가 x만큼 증가
    print(s, end = " ")
```
