### 문제 2. 문제1 변형
- 등록할 이름과아이디 몇개인지 쓰고, 그 갯수 만큼 반복해서 실행하도록 작성해보세요.

- 순서도

image.png

```python
write_count=int(input("이름과 아이디를 몇번 등록하시겠습니까? :"))

for i in range(0,write_count):
    print("%d 번째 이름과 아이디를 등록하세요." % (i+1))
    in_name = input("이름 : ")
    in_id = input("아이디 : ")
    with open("회원명단.txt", "a") as f : # 파일 끝에 추가("a" 모드)
        f.writelines(in_name + " " + in_id + "\n")
    print("%d번째 등록되었습니다."%(i+1))
```

### 문제 3.회원정보 확인 프로그램 만들기
- 회원명단.txt을 못찾으면 "파일을 확인하세요.라고 출력하세요.
- 이름과 아이디를 입력하면 회원정보파일(회원명단.txt)에 등록되어 있는지 찾아서 
- 회원이면 "안녕하세요. 이름님"을 출력하고, 
- 없으면 "이름과 아이디를 확인하세요"를 출력하세요. 



- 순서도
image.png

```python
try :
    f = open("회원명단.txt", "r") 
except :
    print("파일을 확인하세요.")
else :
    in_name = input("이름 입력 : ")
    in_id = input("아이디 입력 : ")

    r_flag = 0  # 이름과 아이디를 찾으면 1로 변경
    
    for data in f :
        data = data.rstrip() # 읽은 문자열의 오른쪽에 있는 '\n'을 제거
        items = data.split() # 공백을 기준으로 잘라서 리스트에 저장(['id', 'pwd'])
        
        if items[0] == in_name and items[1] == in_id : # 값이 일치하면 반복 종료
            print("안녕하세요, %s 님." % in_name)
            r_flag = 1  # 일치하는 이름과 아이디를 찾음
            break

    f.close()
    if r_flag == 0 : # 일치하는 이름과 아이디가 없음
        print("이름과 아이디를 확인하세요.")
```
