## 2.결측값제거(dropna())

* 가.Listwise 방식: 
목록삭제- df.dropna(axis = 0, how = 'any')
 - axis = 0 : 행
 - axis = 1 : 열(컬럼)
 - how = 'any' : 결측치가 있으면 전부 삭제

```python
df.dropna(axis= 0,how='any')
```

```python
df
```

```python
df.dropna(axis= 1,how='any')
```

* 나.Pairwise 방식: 선택하여 삭제


* a.how = 'all' : 전체 행이 결측치이면 삭제

```python
df.dropna(how='all')
```

* b.임계치 설정해서 제거하기
 - thresh = N : 1개의 행(row)에 대하여 최소 N개 이상의 값을 가지는 행만 나옴.

```python
df
```

```python
df.dropna(thresh = 11)
```

* c.특정열의 결측치만 처리
 -subset = ['컬럼명','컬럼명'] : 해당 컬럼의 결측치만 지움.

```python
df.dropna(subset=['SO2','PM10'])
```

## 3.결측값채우기(fillna())

- 결측값을 특정 값으로 채우기 : df.fillna(특정값)
- 결측값을 결측값의 앞 행의 값으로 채우기 : df.fillna(method='ffill') or df.fillna(method='pad')
- 결측값을 결측값의 뒷 행의 값으로 채우기 : df.fillna(method='bfill') or df.fillna(method='backfill')
- 결측값을 각 열의 평균 값으로 채우기 : df.fillna(df.mean())

```python
df.fillna(-999)
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(df.mean())
```

```python
df.fillna(df.var())
```

```python
df['NO2'].value_counts()
```

```python
df['NO2']=df['NO2'].fillna(0.036)
```

```python
df
```

```python
df.fillna(df.mean(),inplace = True)
df
```
