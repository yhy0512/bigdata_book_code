# **파이썬이란?**
* 1990년 암스테르담의 귀도 반 로섬이 개발한 인터프리터 방식의 프로그래밍 언어
 * 귀도 반 로섬 (Guido van Rossum) 이 좋아하던 코미디 프로그램인 Monty Python’ s Flying Circus에서 따온 것
 * 고대 신화에서 파르나소스 산 동굴에 살던 큰 뱀을 일컫는 말이라 파이썬 로고는 뱀

* Linux 및 macOS를 비롯한 많은 Unix 변형과 Windows에서 실행
* Python으로 작성된 프로그램은 일반적으로 다음과 같은 몇 가지 이유로 동등한 C, C++ 또는 Java 프로그램보다 훨씬 짧습니다.
 * 고급 데이터 유형을 사용하면 단일 명령문으로 복잡한 작업을 표현할 수 있습니다.
 * 문 그룹화는 시작 및 끝 괄호 대신 들여쓰기로 수행됩니다.
 * 변수 또는 인수 선언이 필요하지 않습니다.
* 파이썬은 쉽게 배울 수 있고 오픈 소스로 제공되어 무료로 사용할 수 있어 빅데이터 분석에 사용함


# 변수와 객체

* **변수** 
 * 값을 저장하는 메모리 공간
 * 파이썬에서는 변수를 미리 선언하지 않음
 * 변수에 저장해서 사용하는 값의 자료형으로 변수의 자료형이 결정

* **객체(V,value)** 
 * 변수 형태의 속성과 함수 형태의 메서드를 가진 것
 * 각 객체는 자기 의 속성(a,attribute)(내부 데이터)과 메서드(내부 연산)를 가짐
 * 타 프로그래밍 언어와 달리 파이썬에서는 모든 변수와 자료형이 객체로 되어 있음

 * **키워드**
  * 프로그래밍 언어에서만 사용할수 있도록 의미가 부여된 단어

#### 6) 리스트 메서드
- V.append(a) : a를 맨마지막에 추가
- V.sort() : 값의 올림차순으로 인덱스 정렬
- V.reverse() : 값의 내림차순으로 인덱스 정렬

- V.insert(i, a) : i 인덱스에 a를 추가
- V.remove(a) : a값을 제거
- V.pop(i) : i 인덱스 제거,i가 없을 경우 마지막 인덱스 제거
- V.count(a) : a가 몇 개 들어 있는지 반환
